```python
counts = df['DevType'].value_counts()
percent = df['DevType'].value_counts(normalize=True)
percent100 = df['DevType'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%'
pd.DataFrame({'counts': counts, 'per': percent, 'per100': percent100})
```
